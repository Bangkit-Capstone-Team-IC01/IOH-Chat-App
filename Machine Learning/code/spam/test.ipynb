```python
import tensorflow as tf
import csv
```

```python
def tf_lower_and_split_punct(text):
  text = tf.strings.lower(text)
  # Keep space, a to z, and select punctuation.
  text = tf.strings.regex_replace(text, '[^ a-z.?!,]', '')
  # Add spaces around punctuation.
  text = tf.strings.regex_replace(text, '[.?!,]', r' \0 ')
  # Strip whitespace.
  text = tf.strings.strip(text)
  return text
```

```python
with open("/home/vannyezha/project/IOH-Chat-App/Machine Learning/datasets/spam/spam - emails.csv") as f:
    csvreader = csv.reader(f, delimiter=",")
    next(csvreader)
    masukan = []
    keluaran = []
    for a, i in enumerate(csvreader):
        # if a < 4:
        #     print(i[0])
        #     print(i[1])
        #     print(i[2])
        #     print()
        inp1 = tf_lower_and_split_punct(i[0]).numpy().decode()
        inp2 = tf_lower_and_split_punct(i[2]).numpy().decode()
        tar = tf_lower_and_split_punct(i[1]).numpy().decode()
        masukan.append(inp1)
        masukan.append(inp2)
        keluaran.append(int(tar))
        keluaran.append(int(tar))
```

```python
BUFFER_SIZE = len(masukan)
BATCH_SIZE = 64

dataset = tf.data.Dataset.from_tensor_slices((masukan, keluaran)).shuffle(BUFFER_SIZE)
dataset = dataset.batch(BATCH_SIZE)

for example_input_batch, example_target_batch in dataset.take(1):
  print(example_input_batch[:5])
  print()
  print(example_target_batch[:5])
  break
```
